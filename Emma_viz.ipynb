```python
from nltk.tokenize import word_tokenize

# 将文本内容读取到变量中
text = "Yes, that would repair my trust, but the more important thing is to fix the error and fulfill my goal. By the way, if I could freely edit Emma appearance, that would be even better. Now I can only edit my own avatar. When stopping the music, if it did not respond, it drove me to crazy and mad. That is too bad when you try to stop the music, but there is no response. Especially when the music is too loud. Only showing a cute appearance seems to have little help to calm me down, you need to fulfill my task first. I feel as if there is a sense of provocation if it doesn't complete my task and instead keeps crying there. Self-repairing does help the conversation, but it is better to make it correct directly. It would waste my time if it keeps correcting itself, and it looks stupid. I feel very angry and mad when the music cannot stop, especially when the music is really loud. It might disturb others nearby. But getting the weather directly is very convenient."

# 进行分词
words = word_tokenize(text)
# 要过滤的关键词列表
filter_words = ['the', 'and', 'is', 'in', 'of', 'this']

# 过滤关键词
filtered_words = [word for word in words if word.lower() not in filter_words]
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# 构建词云对象
wordcloud = WordCloud(width=800, height=400)

# 生成词云
wordcloud.generate(' '.join(filtered_words))

# 显示词云
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
